### Graphical Analysis of College Major Dataset

```python
%matplotlib inline
import pandas as pd
import matplotlib as plt

recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.head(5))
```

```python
recent_grads.describe()
```

To clean the data we will remove bad rows (rows with null values) using the `df.dropna()` function

```python
raw_data_count = len(recent_grads)
raw_data_count
```

```python
recent_grads = recent_grads.dropna()
raw_data_count = len(recent_grads)
raw_data_count
```

Given the length of the cleaned dataframe, there was only one row with a null value which leaves us will a still full and accurate dataset. 

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', 
                  kind='scatter', title='Unemployment Rate vs. Sample size', 
                  )
```

```python
recent_grads.plot(x='Sample_size', y='Median', 
                  kind='scatter', title='Employed vs. Sample_size', 
                  )
```

```python
recent_grads.plot(x='Full_time', y='Median', 
                  kind='scatter', title='Full Time and Median Salaries', 
                  )
```

```python
recent_grads.plot(x='Part_time', y='Median', 
                  kind='scatter', title='Part Time and Median Salaries', 
                  )
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', 
                  kind='scatter', title='Major Unemplyment Rates Based On Sharewomen', 
                  )
```

```python
recent_grads.plot(x='ShareWomen', y='Median', 
                  kind='scatter', title='Sharewomen Majors and Median Salaries', 
                  )
```

```python
recent_grads.plot(x='Men', y='Median', 
                  kind='scatter', title='Median Salaries of Men', 
                  )
```

```python
recent_grads.plot(x='Women', y='Median', 
                  kind='scatter', title='Median Salaries of Women', 
                  )
```

```python
recent_grads.plot(x='Total', y='Median', 
                  kind='scatter', title='Median Salaries and Popularities' 
                  )
```

Interpretation of The Above Plots:
- Women and men have similar median salaries to men but the unemployment rate is higher in majors with a higher share of women. The median salaries for majors that are predominantly women have the lowest median salaries. This means indicates that working women make similar amounts to men but the data is skewed toward men making more money because more women are unemployed. The remainder of the pay gap is evedently a result of more women pursuing lower paying industries.
- There is no significant correlation between major popularity and median salaries but outliers with the highest salaries are in rare major categories. 
- Part time employees make only slightly less than full time employees on average.

```python
from pandas.plotting import scatter_matrix

scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

Insights from histogram and scatterplot matrix:
- It's evident that sample size and median salaries have no strong relationship except that outlier (rare) majors are the only ones in the highest salary ranges. 
- Unemployment rates are not correlated with sample size but all the outliers (majors with highest and lowest unemployment rate) always have a small sample size. The scatter plot for sample size and unemployment rate resembles a normal distribution. 

```python
recent_grads[:5].plot.bar(x='Major', y='ShareWomen', legend=False)
```

```python
recent_grads[167:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

The above bar graphs clearly indicate that majors with the highest median salaries are dominated by men and majors with the lowest median salaries are dominated by women. 

```python
gender_majors = recent_grads.plot.bar(x='Major', y='ShareWomen', legend=False)
gender_majors.tick_params(bottom='off', left='off', right='off')
gender_majors.set_xticklabels('0', rotation=90)
gender_majors.set_xlabel('Major Salaries (left=highest)')
gender_majors.set_ylabel('Share of Women')
```

The above bar chart shows a strong trend toward men chosing highing pying majors and women choosing lower paying majors.
